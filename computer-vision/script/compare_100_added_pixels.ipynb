```python
! pip install opencv-python numpy matplotlib
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
```

## Grayscale Image

```python
url = 'https://www.inf.ufrgs.br/~cwaraujo/inf01046/trabalho%20pr%C3%A1tico%201%20-%20relat%C3%B3rio_arquivos/image024.jpg'
with urllib.request.urlopen(url) as url_response:
    img_array = np.array(bytearray(url_response.read()), dtype=np.uint8)
    gray_img = cv2.imdecode(img_array, cv2.IMREAD_GRAYSCALE)
```

### Original
After executing the code, the image will be displayed in a new window.

```python
cv2.imshow('Grayscale Image', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### With 100 pixels added
After executing the code, the image will be displayed in a new window.

```python
gray_img_with_100_pixels_added = cv2.add(gray_img, 100)

cv2.imshow('Grayscale image with 100 pixels added', gray_img_with_100_pixels_added)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Comparing the images

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(gray_img, cmap='gray')
axs[0].set_title('Original')
axs[1].imshow(gray_img_with_100_pixels_added, cmap='gray')
axs[1].set_title('With 100 pixels added')
plt.show()
```

## Color Image

## Conclusion

When you add 100 units to each pixel of a grayscale image, the maximum pixel value can exceed the limit of 255, which is the maximum value for an 8-bit pixel. This means that some pixels that originally had different values may end up having the same value after adding 100 units. As a result, the resulting image will have less detail since some pixels become less distinct from each other. Additionally, this operation can also cause pixels that originally had similar values to become very different, which can lead to a loss of sharpness in the image
